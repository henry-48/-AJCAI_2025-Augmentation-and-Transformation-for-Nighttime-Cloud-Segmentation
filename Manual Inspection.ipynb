```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

# manual inspcetion analyze 

```python
log_f = pd.read_csv("./button_press_log.csv",index_col=0)

log_f['combined'] = log_f.apply(lambda row: ' '.join(row.dropna().tolist()), axis=1)
```

```python
def str_to_ls(input_string):
    list_of_strings = input_string.split("] [")

    # Adjust the brackets for each element to make them valid lists
    list_of_strings = [s.replace("]", "").replace("[", "").strip() for s in list_of_strings]

    # Convert each list-like string into an actual Python list using ast.literal_eval
    list_of_lists = [ast.literal_eval(f"[{s}]") for s in list_of_strings]
    return list_of_lists
```

```python
log_f['combined_list'] = log_f['combined'].apply(str_to_ls)
comb = log_f['combined_list']
```

```python
agreement = ['Agree','Disagree','Partially agree']

agree_dict = {key: 0 for key in agreement}

clear_dcit = {key: 0 for key in agreement}

few_cloud = {key: 0 for key in agreement}

most_cloud = {key: 0 for key in agreement}

full_cloud = {key: 0 for key in agreement}
index = log_f.index.tolist()
for i in range(len(comb)):
    choice = comb[i][-1]
    
    agree, cloud = choice
    
    agree_dict[agree] += 1
    
    if cloud == "Full Cloudy":
        
        full_cloud[agree] += 1
    elif cloud == "Mostly Cloudy":
        print(index[i])
        most_cloud[agree] += 1
    elif cloud == "Few Cloudy":
        
        few_cloud[agree] += 1
    elif cloud == "Clear":
        clear_dcit[agree] += 1
    else:
        continue
        
```

```python
plt.bar(agree_dict.keys(), agree_dict.values(), width=0.8, edgecolor='black')
plt.title("Histogram of Overall Manual Inspection")
plt.ylabel("Occurance")
plt.show()
```

```python
plt.bar(full_cloud.keys(), full_cloud.values(), width=0.8, edgecolor='black')
plt.title("Histogram of Overall Manual Inspection in Fully Cloudy Conditions")
plt.ylabel("Occurance")
plt.show()
```
